```python
# Pearson Correlation
from scipy import stats
```

```python
pearson_coef, p_value = stats.pearsonr(df["weight"], df["Kilometer_per_liter"])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
# Since the p-value is  <  0.001, the correlation between weight and Kilometer_per_liter is statistically significant, and the 
# linear relationship is quite strong
```

```python
pearson_coef, p_value = stats.pearsonr(df["displacement"], df["Kilometer_per_liter"])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
# Since the p-value is  <  0.001, the correlation between displacement and Kilometer_per_liter is statistically significant, and the 
# linear relationship is quite strong
```

```python
pearson_coef, p_value = stats.pearsonr(df["origin"], df["Kilometer_per_liter"])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
# Since the p-value is  <  0.001, the correlation between origin and Kilometer_per_liter is statistically significant, and the 
# linear relationship is moderate
```

```python
pearson_coef, p_value = stats.pearsonr(df["horsepower"], df["Kilometer_per_liter"])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
# Since the p-value is  <  0.001, the correlation between horsepower and Kilometer_per_liter is statistically significant, and the 
# linear relationship is strong
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df['name'] = le.fit_transform(df.name.values)
```

```python
df.shape
```

```python
#Feature Scaling

data = preprocessing.StandardScaler().fit_transform(df)
df2 = pd.DataFrame(data)
print(df2.head())
```

```python
y_data = df.iloc[:,8:]
x_data = df.iloc[:,:8]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.20, random_state=1) 
```

```python
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
regr = SGDRegressor()
```

```python
regr.fit(x_train, y_train)
```

```python
print(regr.coef_)
```

```python
print(regr.intercept_)
```

```python
print('The r-squared value is: ', regr.score(x_train,y_train))
```

```python
y_pred = regr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
print('The mean square error of price and predicted value is: ', mse)
```

```python
print(r2_score(y_test, y_pred))
```
